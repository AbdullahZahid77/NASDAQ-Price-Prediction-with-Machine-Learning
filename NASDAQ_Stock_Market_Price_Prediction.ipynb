```python
del nasdaq["Dividends"]
del nasdaq["Stock Splits"]
#not needed to predict the stock market
```

```python
nasdaq["price_tomorrow"]= nasdaq["Close"].shift(-1)
#shifted close column values one step behind to predict the price for tomorrow
```

```python
 nasdaq['Target']=(nasdaq["price_tomorrow"]>nasdaq['Close']).astype(int)
 nasdaq
 #we will predict the target value
 #convert the BOOL values to INT to be used for machine learning
```

```python
nasdaq=nasdaq.loc['1990-01-01':].copy()
#removing the very old data as it could cause predictions to come out wrong
```

```python
from sklearn.ensemble import RandomForestClassifier
#machine learning model
```

```python
model=RandomForestClassifier(n_estimators=1000,min_samples_split=100,random_state=1)
```

```python
train=nasdaq.iloc[:-100]  #start to -100 rows
test=nasdaq.iloc[-100:]   #the last 100 rows
```

```python
predictors=['Close','Volume','Open', 'High','Low'] #predcitors for my model
```

```python
model.fit(nasdaq[predictors],nasdaq['Target'])
```

```python
from sklearn.metrics import precision_score
```

```python
predictions=model.predict(test[predictors]) #generating predictions on the train data set
```

```python
predictions
```

```python
import pandas as pd
predictions=pd.Series(predictions,index=test.index)
```

```python
predictions
```

```python
probablity=(precision_score(test['Target'],predictions))*100
probablity
#the predictions are 80% correct
```

```python
conc=pd.concat([test['Target'],predictions],axis=1)
conc
```

```python
def predicting (model,test,train,predictors): #prediction function
  model.fit(nasdaq[predictors],nasdaq['Target'])
  predictions=model.predict(test[predictors]) #generating predictions on the train data set
  predictions=pd.Series(predictions,index=test.index,name="Predictions")
  conc=pd.concat([test['Target'],predictions],axis=1)
  return conc
```

```python
def backtesting (model,nasdaq,predictors,start=2500,step=250): #backtesting function
    all_prediction=[]
    for i in range(start,nasdaq.shape[0],step):
      train=nasdaq.iloc[0:i].copy()
      test=nasdaq.iloc[i:(i+step)].copy()
      predictions=predicting(model,test,train,predictors)
      all_prediction.append(predictions)
      return pd.concat(all_prediction)

```

```python
predictions=backtesting(model,nasdaq,predictors)
```

```python
predictions['Predictions'].value_counts()
```

```python
(precision_score(predictions['Target'],predictions['Predictions']))*100
```

```python
def predicting1 (model,test,train,predictors): #prediction function
  model.fit(nasdaq[predictors],nasdaq['Target'])
  predictions=model.predict_proba(test[predictors])[:,1] #generating probablities on the train data set
  predictions[predictions>=0.5]=1
  predictions[predictions<0.5]=0
  predictions=pd.Series(predictions,index=test.index,name="Predictions")
  conc=pd.concat([test['Target'],predictions],axis=1)
  return conc
```
